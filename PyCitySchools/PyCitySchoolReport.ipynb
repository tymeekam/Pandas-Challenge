## PYCITY SCHOOLS ANALYSIS

We analyzed 15 schools within the PyCity School District both Charter and District type by comparing their overal math and reading passing scores. Overall 65% of all students within the Pycity Schools are passing both categories. As we analyzed the budget of each school type it was determined that having a higher budget of more than 630.00 dollars did not neccessarily result in better math or reading skills overall, but in fact, the high budgeted schools under performed compared to the schools with a lower bugeted range of 585.00-630.00 dollars.

This appeared relative to the size of the schools. As the medium and small sized schools, those with 2000 or less students, outperformed the schools with relatively larger, of 2000-5000 with a passing average of 89% and higher.

Overall we were able to determine that as a whole, the charter schools within this district has a higher overall passing rate in both math and reading compared to those of the district type based schools. We can dive deeper into these numbers and determine if the smaller class sizes, location of schools and overall health of students mental and well beging effect these outcomes as a whole.

# MATH SCORE BY GRADE

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]
```

```python
# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grade_math_scores = ninth_graders.groupby("school_name")["math_score"].mean().reset_index()
tenth_grader_math_scores = tenth_graders.groupby("school_name")["math_score"].mean().reset_index()
eleventh_grader_math_scores = eleventh_graders.groupby("school_name")["math_score"].mean().reset_index()
twelfth_grader_math_scores = twelfth_graders.groupby("school_name")["math_score"].mean().reset_index()

#  Combine each of the scores above into single DataFrame called `math_scores_by_grade`
# math_scores_by_grade = 
math_scores_by_grade = pd.merge(ninth_grade_math_scores, tenth_grader_math_scores, on="school_name", suffixes=('_9th', '_10th'))
math_scores_by_grade = pd.merge(math_scores_by_grade, eleventh_grader_math_scores, on="school_name")
math_scores_by_grade = pd.merge(math_scores_by_grade, twelfth_grader_math_scores, on="school_name", suffixes=('_11th', '_12th'))

math_scores_by_grade.columns = ['', '9th', '10th', '11th', '12th']


# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade
```

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the the `reading_score` column for each.
ninth_grade_reading_scores = ninth_graders.groupby("school_name")["reading_score"].mean().reset_index()
tenth_grader_reading_scores = tenth_graders.groupby("school_name")["reading_score"].mean().reset_index()
eleventh_grader_reading_scores = eleventh_graders.groupby("school_name")["reading_score"].mean().reset_index()
twelfth_grader_reading_scores = twelfth_graders.groupby("school_name")["reading_score"].mean().reset_index()

# Combine each of the scores above into single DataFrame called `reading_scores_by_grade`
reading_scores_by_grade = pd.merge(ninth_grade_reading_scores, tenth_grader_reading_scores, on="school_name", suffixes=('_9th', '_10th'))
reading_scores_by_grade = pd.merge(reading_scores_by_grade, eleventh_grader_reading_scores, on="school_name")
reading_scores_by_grade = pd.merge(reading_scores_by_grade, twelfth_grader_reading_scores, on="school_name", suffixes=('_11th', '_12th'))

reading_scores_by_grade.columns = ['', '9th', '10th', '11th', '12th']

# Minor data wrangling
reading_scores_by_grade.index.name = None

# Display the DataFrame
reading_scores_by_grade
```
